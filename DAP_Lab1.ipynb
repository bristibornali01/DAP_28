```python
#Compute nPr (permutations).


import math

def permutations(n, r):
    if not isinstance(n, int) or not isinstance(r, int):
        return "Inputs n and r must be integers."
    if n < 0 or r < 0:
        return "Inputs n and r must be non-negative."
    if r > n:
        return "r cannot be greater than n for permutations."

    return math.factorial(n) // math.factorial(n - r)

# Example Usage:
print(f"P(5, 2) = {permutations(5, 2)}") # Expected: 5! / (5-2)! = 5! / 3! = 120 / 6 = 20
print(f"P(10, 3) = {permutations(10, 3)}") # Expected: 10! / 7! = 10 * 9 * 8 = 720
print(f"P(4, 4) = {permutations(4, 4)}") # Expected: 4! / 0! = 24 / 1 = 24
print(f"P(7, 0) = {permutations(7, 0)}") # Expected: 7! / 7! = 1
print(f"P(0, 0) = {permutations(0, 0)}") # Expected: 0! / 0! = 1 / 1 = 1
print(f"P(5, 6) = {permutations(5, 6)}") # Expected: Error message
print(f"P(-5, 2) = {permutations(-5, 2)}") # Expected: Error message
```

```python

#Compute nCr (combinations) using factorial + simplification.



import math

def combinations(n, r):
    if not isinstance(n, int) or not isinstance(r, int):
        return "Inputs n and r must be integers."
    if n < 0 or r < 0:
        return "Inputs n and r must be non-negative."
    if r > n:
        return "r cannot be greater than n for combinations."

    # Using the formula nCr = n! / (r! * (n-r)!)
    return math.factorial(n) // (math.factorial(r) * math.factorial(n - r))

# Example Usage:
print(f"C(5, 2) = {combinations(5, 2)}") # Expected: 5! / (2! * 3!) = 10
print(f"C(10, 3) = {combinations(10, 3)}") # Expected: 10! / (3! * 7!) = 120
print(f"C(4, 4) = {combinations(4, 4)}") # Expected: 4! / (4! * 0!) = 1
print(f"C(7, 0) = {combinations(7, 0)}") # Expected: 7! / (0! * 7!) = 1
print(f"C(0, 0) = {combinations(0, 0)}") # Expected: 0! / (0! * 0!) = 1
print(f"C(5, 6) = {combinations(5, 6)}") # Expected: Error message
print(f"C(-5, 2) = {combinations(-5, 2)}") # Expected: Error message
```

```python

# Find LCM using GCD.


def find_lcm(a, b):
    # Ensure a and b are non-negative for LCM calculation validity
    # The formula LCM(a,b) = |a*b| / GCD(a,b) is generally used.
    # Handle edge case where either number is zero (LCM is 0)
    if a == 0 or b == 0:
        return 0

    # Make sure to use the absolute values for GCD calculation to handle negative inputs gracefully
    abs_a = abs(a)
    abs_b = abs(b)

    # Call the existing gcd_euclid function
    common_divisor = gcd_euclid(abs_a, abs_b)

    # Use the formula: LCM(a, b) = (a * b) / GCD(a, b)
    # Using integer division for the result, as LCM should be an integer.
    return (abs_a * abs_b) // common_divisor

# Example Usage:
print(f"LCM of 4 and 6 is: {find_lcm(4, 6)}")       # Expected: 12
print(f"LCM of 15 and 20 is: {find_lcm(15, 20)}")     # Expected: 60
print(f"LCM of 7 and 11 is: {find_lcm(7, 11)}")     # Expected: 77 (coprime)
print(f"LCM of 12 and 4 is: {find_lcm(12, 4)}")      # Expected: 12
print(f"LCM of 0 and 5 is: {find_lcm(0, 5)}")      # Expected: 0
print(f"LCM of 7 and 0 is: {find_lcm(7, 0)}")      # Expected: 0
print(f"LCM of -4 and 6 is: {find_lcm(-4, 6)}")     # Expected: 12
print(f"LCM of 0 and 0 is: {find_lcm(0, 0)}")      # Expected: 0
```

```python
# Find HCF (GCD) of three numbers using the existing gcd_euclid function.

def find_hcf_three_numbers(a, b, c):
    # HCF(a, b, c) = HCF(a, HCF(b, c))
    # Ensure numbers are non-negative for GCD calculation validity
    abs_a = abs(a)
    abs_b = abs(b)
    abs_c = abs(c)

    # Handle case where all numbers are zero
    if abs_a == 0 and abs_b == 0 and abs_c == 0:
        return 0 # Conventionally, HCF(0,0,0) is 0

    # Calculate HCF of b and c first
    hcf_bc = gcd_euclid(abs_b, abs_c)

    # Then calculate HCF of a and the result of hcf_bc
    return gcd_euclid(abs_a, hcf_bc)

# Example Usage:
print(f"HCF of 12, 18, 30 is: {find_hcf_three_numbers(12, 18, 30)}") # Expected: 6
print(f"HCF of 10, 15, 20 is: {find_hcf_three_numbers(10, 15, 20)}") # Expected: 5
print(f"HCF of 7, 11, 13 is: {find_hcf_three_numbers(7, 11, 13)}")   # Expected: 1 (coprime)
print(f"HCF of 0, 5, 10 is: {find_hcf_three_numbers(0, 5, 10)}")   # Expected: 5
print(f"HCF of -12, 18, -30 is: {find_hcf_three_numbers(-12, 18, -30)}") # Expected: 6
print(f"HCF of 0, 0, 0 is: {find_hcf_three_numbers(0, 0, 0)}")       # Expected: 0
```

```python
# Compute power a^b using loop (no **).

def power(base, exponent):
    if not isinstance(base, (int, float)) or not isinstance(exponent, int):
        return "Base must be a number and exponent must be an integer."

    if exponent == 0:
        return 1

    result = 1
    if exponent > 0:
        for _ in range(exponent):
            result *= base
    else: # exponent < 0
        for _ in range(abs(exponent)):
            result *= base
        result = 1 / result
    return result

# Example Usage:
print(f"2^3 = {power(2, 3)}")    # Expected: 8
print(f"5^0 = {power(5, 0)}")    # Expected: 1
print(f"3^4 = {power(3, 4)}")    # Expected: 81
print(f"2^-2 = {power(2, -2)}")  # Expected: 0.25
print(f"10^1 = {power(10, 1)}") # Expected: 10
print(f"0^5 = {power(0, 5)}")    # Expected: 0
print(f"(-2)^3 = {power(-2, 3)}") # Expected: -8
print(f"(-2)^4 = {power(-2, 4)}") # Expected: 16
print(f"2.5^2 = {power(2.5, 2)}") # Expected: 6.25
print(f"2^a = {power(2, 'a')}")  # Expected: Error message
```

```python
#Reverse a number (no string).

def reverse_number(n):
    reversed_num = 0
    while n > 0:
        digit = n % 10  # Get the last digit
        reversed_num = reversed_num * 10 + digit  # Append the digit to the reversed number
        n = n // 10  # Remove the last digit
    return reversed_num

# Example usage:
number_to_reverse = 12345
reversed_result = reverse_number(number_to_reverse)

print(f"Original number: {number_to_reverse}")
print(f"Reversed number: {reversed_result}")

number_to_reverse_2 = 9870
reversed_result_2 = reverse_number(number_to_reverse_2)

print(f"Original number: {number_to_reverse_2}")
print(f"Reversed number: {reversed_result_2}")
```

```python
# Compute power a^b mod m (fast exponentiation).

def power_mod(base, exponent, modulus):
    if not isinstance(base, int) or not isinstance(exponent, int) or not isinstance(modulus, int):
        return "Inputs base, exponent, and modulus must be integers."
    if modulus <= 0:
        return "Modulus must be a positive integer."
    if exponent < 0:
        return "Exponent cannot be negative for modular exponentiation."

    result = 1
    base %= modulus  # Ensure base is within the modulus range

    while exponent > 0:
        if exponent % 2 == 1:  # If exponent is odd, multiply result with base
            result = (result * base) % modulus
        base = (base * base) % modulus  # Square the base
        exponent //= 2  # Halve the exponent

    return result

# Example Usage:
print(f"(2^3) mod 5 = {power_mod(2, 3, 5)}")     # Expected: (8) mod 5 = 3
print(f"(3^4) mod 7 = {power_mod(3, 4, 7)}")     # Expected: (81) mod 7 = 4
print(f"(10^0) mod 13 = {power_mod(10, 0, 13)}")   # Expected: (1) mod 13 = 1
print(f"(7^1) mod 2 = {power_mod(7, 1, 2)}")       # Expected: (7) mod 2 = 1
print(f"(4^13) mod 497 = {power_mod(4, 13, 497)}") # Expected: 445 (a larger example)
print(f"(2^-2) mod 5 = {power_mod(2, -2, 5)}")    # Expected: Error message
```

```python
#Compute sin(x) using Taylor series (n terms).


import math

def taylor_sin_x(x, n_terms):
    if n_terms <= 0:
        return "Number of terms must be positive"

    sin_x_approx = 0
    for k in range(n_terms):
        # Term for sin(x) Taylor series is (-1)^k * x^(2k+1) / (2k+1)!
        term = ((-1)**k) * (x**(2*k + 1)) / math.factorial(2*k + 1)
        sin_x_approx += term
    return sin_x_approx

# Example Usage:
# sin(0) (should be 0)
print(f"sin(0) with 5 terms: {taylor_sin_x(0, 5)}")

# sin(math.pi/2) (should be approx 1)
print(f"sin(pi/2) with 10 terms: {taylor_sin_x(math.pi/2, 10)}")
print(f"sin(pi/2) with 20 terms: {taylor_sin_x(math.pi/2, 20)}")

# sin(math.pi) (should be approx 0)
print(f"sin(pi) with 10 terms: {taylor_sin_x(math.pi, 10)}")
print(f"sin(pi) with 20 terms: {taylor_sin_x(math.pi, 20)}")

# Compare with math.sin()
actual_sin_pi_over_2 = math.sin(math.pi/2)
print(f"Actual sin(pi/2): {actual_sin_pi_over_2}")

actual_sin_pi = math.sin(math.pi)
print(f"Actual sin(pi): {actual_sin_pi}")
```

```python
# Find prime factors of a number.

def prime_factors(n):
    factors = []
    # Handle case for n < 2
    if n < 2:
        return factors

    # Handle factor 2
    while n % 2 == 0:
        factors.append(2)
        n //= 2

    # Handle odd factors from 3 upwards
    i = 3
    while i * i <= n:
        while n % i == 0:
            factors.append(i)
            n //= i
        i += 2

    # If n is still greater than 2, it must be a prime factor itself
    if n > 2:
        factors.append(n)

    return factors

# Example Usage:
print(f"Prime factors of 12: {prime_factors(12)}")   # Expected: [2, 2, 3]
print(f"Prime factors of 30: {prime_factors(30)}")   # Expected: [2, 3, 5]
print(f"Prime factors of 7: {prime_factors(7)}")     # Expected: [7]
print(f"Prime factors of 1: {prime_factors(1)}")     # Expected: []
print(f"Prime factors of 0: {prime_factors(0)}")     # Expected: []
print(f"Prime factors of -10: {prime_factors(-10)}") # Expected: [] (or handle as per definition)
print(f"Prime factors of 100: {prime_factors(100)}") # Expected: [2, 2, 5, 5]
print(f"Prime factors of 999: {prime_factors(999)}") # Expected: [3, 3, 3, 37]
```

```python
#Check palindrome number (no string).


def is_palindrome_number(n):
    if n < 0:  # Negative numbers are not palindromes
        return False
    if n == 0:
        return True

    original_num = n
    reversed_num = 0
    while n > 0:
        digit = n % 10
        reversed_num = reversed_num * 10 + digit
        n //= 10

    return original_num == reversed_num

# Example usage:
print(f"Is 121 a palindrome? {is_palindrome_number(121)}")
print(f"Is 12321 a palindrome? {is_palindrome_number(12321)}")
print(f"Is 12345 a palindrome? {is_palindrome_number(12345)}")
print(f"Is 1001 a palindrome? {is_palindrome_number(1001)}")
print(f"Is 0 a palindrome? {is_palindrome_number(0)}")
print(f"Is -121 a palindrome? {is_palindrome_number(-121)}")
```

```python
# Check prime number (optimized up to √n).

import math

def is_prime(n):
    if n < 2:
        return False
    if n == 2 or n == 3:
        return True
    if n % 2 == 0 or n % 3 == 0:
        return False

    i = 5
    while i * i <= n:
        if n % i == 0 or n % (i + 2) == 0:
            return False
        i += 6
    return True

# Example Usage:
print(f"Is 2 a prime number? {is_prime(2)}")     # Expected: True
print(f"Is 7 a prime number? {is_prime(7)}")     # Expected: True
print(f"Is 11 a prime number? {is_prime(11)}")    # Expected: True
print(f"Is 10 a prime number? {is_prime(10)}")    # Expected: False
print(f"Is 1 a prime number? {is_prime(1)}")     # Expected: False
print(f"Is 0 a prime number? {is_prime(0)}")     # Expected: False
print(f"Is -5 a prime number? {is_prime(-5)}")    # Expected: False
print(f"Is 29 a prime number? {is_prime(29)}")    # Expected: True
print(f"Is 49 a prime number? {is_prime(49)}")    # Expected: False (7*7)
print(f"Is 97 a prime number? {is_prime(97)}")    # Expected: True
```

```python
##Check Armstrong number (general n-digit).


def is_armstrong_number(n):
    if n < 0:
        return False
    if n == 0:
        return True

    original_num = n
    num_digits = 0
    temp_num = n
    while temp_num > 0:
        temp_num //= 10
        num_digits += 1

    sum_of_powers = 0
    temp_num = n
    while temp_num > 0:
        digit = temp_num % 10
        sum_of_powers += digit ** num_digits
        temp_num //= 10

    return sum_of_powers == original_num

# Example usage:
print(f"Is 153 an Armstrong number? {is_armstrong_number(153)}")  # 1^3 + 5^3 + 3^3 = 1 + 125 + 27 = 153
print(f"Is 370 an Armstrong number? {is_armstrong_number(370)}")  # 3^3 + 7^3 + 0^3 = 27 + 343 + 0 = 370
print(f"Is 371 an Armstrong number? {is_armstrong_number(371)}")  # 3^3 + 7^3 + 1^3 = 27 + 343 + 1 = 371
print(f"Is 1634 an Armstrong number? {is_armstrong_number(1634)}") # 1^4 + 6^4 + 3^4 + 4^4 = 1 + 1296 + 81 + 256 = 1634
print(f"Is 123 an Armstrong number? {is_armstrong_number(123)}")
print(f"Is 9 an Armstrong number? {is_armstrong_number(9)}")
print(f"Is 0 an Armstrong number? {is_armstrong_number(0)}")
print(f"Is -153 an Armstrong number? {is_armstrong_number(-153)}")
```

```python
# Print primes in range using Sieve of Eratosthenes.

def sieve_of_eratosthenes(n):
    if n < 2:
        return []

    # Create a boolean array "prime[0..n]" and initialize
    # all entries it as true. A value in prime[i] will
    # finally be false if i is Not a prime, else true.
    prime = [True for i in range(n + 1)]
    p = 2
    while (p * p <= n):
        # If prime[p] is still true, then it is a prime
        if (prime[p] == True):
            # Update all multiples of p
            for i in range(p * p, n + 1, p):
                prime[i] = False
        p += 1

    # Collect all prime numbers
    primes_in_range = []
    for p in range(2, n + 1):
        if prime[p]:
            primes_in_range.append(p)
    return primes_in_range

# Example Usage:
print(f"Primes up to 10: {sieve_of_eratosthenes(10)}")     # Expected: [2, 3, 5, 7]
print(f"Primes up to 20: {sieve_of_eratosthenes(20)}")     # Expected: [2, 3, 5, 7, 11, 13, 17, 19]
print(f"Primes up to 2: {sieve_of_eratosthenes(2)}")       # Expected: [2]
print(f"Primes up to 1: {sieve_of_eratosthenes(1)}")       # Expected: []
print(f"Primes up to 0: {sieve_of_eratosthenes(0)}")       # Expected: []
print(f"Primes up to 50: {sieve_of_eratosthenes(50)}")
```

```python
#Check Perfect number.


def is_perfect_number(n):
    if n <= 0:
        return False

    sum_divisors = 0
    for i in range(1, n):
        if n % i == 0:
            sum_divisors += i

    return sum_divisors == n

# Example usage:
print(f"Is 6 a perfect number? {is_perfect_number(6)}")    # Divisors: 1, 2, 3. Sum = 1+2+3 = 6
print(f"Is 28 a perfect number? {is_perfect_number(28)}")   # Divisors: 1, 2, 4, 7, 14. Sum = 1+2+4+7+14 = 28
print(f"Is 496 a perfect number? {is_perfect_number(496)}") # Divisors: 1, 2, 4, 8, 16, 31, 62, 124, 248. Sum = 496
print(f"Is 10 a perfect number? {is_perfect_number(10)}")  # Divisors: 1, 2, 5. Sum = 1+2+5 = 8 (not 10)
print(f"Is 1 a perfect number? {is_perfect_number(1)}")
print(f"Is 0 a perfect number? {is_perfect_number(0)}")
print(f"Is -6 a perfect number? {is_perfect_number(-6)}")
```

```python
#Check Strong number (sum of factorial of digits)

def factorial(n):
    if n == 0:
        return 1
    res = 1
    for i in range(1, n + 1):
        res *= i
    return res

def is_strong_number(n):
    if n < 0:
        return False
    if n == 0:
        return False # 0 is not typically considered a strong number

    original_num = n
    sum_of_factorials = 0
    temp_num = n

    while temp_num > 0:
        digit = temp_num % 10
        sum_of_factorials += factorial(digit)
        temp_num //= 10

    return sum_of_factorials == original_num

# Example usage:
print(f"Is 145 a strong number? {is_strong_number(145)}") # 1! + 4! + 5! = 1 + 24 + 120 = 145
print(f"Is 2 a strong number? {is_strong_number(2)}")     # 2! = 2
print(f"Is 1 a strong number? {is_strong_number(1)}")     # 1! = 1
print(f"Is 10 a strong number? {is_strong_number(10)}")   # 1! + 0! = 1 + 1 = 2 (not 10)
print(f"Is 146 a strong number? {is_strong_number(146)}")
print(f"Is 0 a strong number? {is_strong_number(0)}")
print(f"Is -145 a strong number? {is_strong_number(-145)}")
```

```python
# Compute factorial using recursion.

def factorial_recursive(n):
    if not isinstance(n, int) or n < 0:
        return "Input must be a non-negative integer."
    if n == 0:
        return 1
    else:
        return n * factorial_recursive(n - 1)

# Example Usage:
print(f"Factorial of 5: {factorial_recursive(5)}")   # Expected: 120
print(f"Factorial of 0: {factorial_recursive(0)}")   # Expected: 1
print(f"Factorial of 1: {factorial_recursive(1)}")   # Expected: 1
print(f"Factorial of 10: {factorial_recursive(10)}") # Expected: 3628800
print(f"Factorial of -3: {factorial_recursive(-3)}") # Expected: Error message
print(f"Factorial of 3.5: {factorial_recursive(3.5)}") # Expected: Error message
```

```python
#Check Harshad/Niven number.

def is_harshad_number(n):
    if n <= 0:
        return False

    original_num = n
    sum_of_digits = 0
    temp_num = n

    while temp_num > 0:
        digit = temp_num % 10
        sum_of_digits += digit
        temp_num //= 10

    # A Harshad number is divisible by the sum of its digits
    return original_num % sum_of_digits == 0

# Example usage:
print(f"Is 18 a Harshad number? {is_harshad_number(18)}")   # 1+8 = 9, 18 % 9 == 0 (True)
print(f"Is 21 a Harshad number? {is_harshad_number(21)}")   # 2+1 = 3, 21 % 3 == 0 (True)
print(f"Is 156 a Harshad number? {is_harshad_number(156)}") # 1+5+6 = 12, 156 % 12 == 0 (True)
print(f"Is 17 a Harshad number? {is_harshad_number(17)}")   # 1+7 = 8, 17 % 8 != 0 (False)
print(f"Is 0 a Harshad number? {is_harshad_number(0)}")    # False
print(f"Is -18 a Harshad number? {is_harshad_number(-18)}") # False
```

```python
#Check Automorphic number (ends with its square).

def is_automorphic_number(n):
    if n < 0:
        return False

    square = n * n
    # Convert both to strings to compare the ending digits
    return str(square).endswith(str(n))

# Example usage:
print(f"Is 1 a automorphic number? {is_automorphic_number(1)}")    # 1*1 = 1
print(f"Is 5 a automorphic number? {is_automorphic_number(5)}")    # 5*5 = 25
print(f"Is 6 a automorphic number? {is_automorphic_number(6)}")    # 6*6 = 36
print(f"Is 25 a automorphic number? {is_automorphic_number(25)}")  # 25*25 = 625
print(f"Is 76 a automorphic number? {is_automorphic_number(76)}")  # 76*76 = 5776
print(f"Is 23 a automorphic number? {is_automorphic_number(23)}")  # 23*23 = 529 (False)
print(f"Is 0 a automorphic number? {is_automorphic_number(0)}")    # 0*0 = 0
print(f"Is -1 a automorphic number? {is_automorphic_number(-1)}")   # False
```

```python
#Check Neon number (sum digits of square equals number).

def is_neon_number(n):
    if n < 0:
        return False

    square = n * n
    sum_of_digits_square = 0

    # Calculate the sum of digits of the square
    temp_square = square
    while temp_square > 0:
        digit = temp_square % 10
        sum_of_digits_square += digit
        temp_square //= 10

    return sum_of_digits_square == n

# Example usage:
print(f"Is 0 a Neon number? {is_neon_number(0)}")   # 0*0 = 0, sum of digits 0 = 0
print(f"Is 1 a Neon number? {is_neon_number(1)}")   # 1*1 = 1, sum of digits 1 = 1
print(f"Is 9 a Neon number? {is_neon_number(9)}")   # 9*9 = 81, sum of digits 8+1 = 9
print(f"Is 45 a Neon number? {is_neon_number(45)}") # 45*45 = 2025, sum of digits 2+0+2+5 = 9 (not 45)
print(f"Is -9 a Neon number? {is_neon_number(-9)}") # False
```

```python
# Find GCD using Euclid algorithm.

def gcd_euclid(a, b):
    # Ensure a and b are non-negative
    a = abs(a)
    b = abs(b)

    while b:
        a, b = b, a % b
    return a

# Example usage:
print(f"GCD of 48 and 18 is: {gcd_euclid(48, 18)}")   # Expected: 6
print(f"GCD of 101 and 103 is: {gcd_euclid(101, 103)}") # Expected: 1 (coprime)
print(f"GCD of 0 and 5 is: {gcd_euclid(0, 5)}")      # Expected: 5
print(f"GCD of 7 and 0 is: {gcd_euclid(7, 0)}")      # Expected: 7
print(f"GCD of 12 and 12 is: {gcd_euclid(12, 12)}")   # Expected: 12
print(f"GCD of -48 and 18 is: {gcd_euclid(-48, 18)}") # Expected: 6
print(f"GCD of 0 and 0 is: {gcd_euclid(0, 0)}")      # Expected: 0 (conventionally)
```
